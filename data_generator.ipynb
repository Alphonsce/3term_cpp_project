```python
sp = []
for _ in range(500):
    sp.append(
    tuple(np.random.uniform(low=-10, high=10, size=2))
    )
```

```python
y = []
for elem in sp:
    if np.sign(elem[0]) == np.sign(elem[1]):
        y.append(1.)
    else:
        y.append(0.)
```
